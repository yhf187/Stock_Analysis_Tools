# Stock Analysis Module

```python
def run_main():
    while True:
            print(str(datetime.now()))
            symbols = ['TSLA', 'NIO', 'XPEV', 'LI',
                       'PYPL', 'AMZN', 'EBAY', 'ABNB', 'DKNG', 'DASH',
                       'FB', 'GOOG', 'ZM', 'AAPL', 'MSFT', 'UBER', 'LYFT', 'CRM', 'SNOW', 'VLDR',
                       'BABA', 'PDD', 'JD', 'TCEHY',
                       'BA', 'DAL', 'UAL', 'SPG', 'CCL', 'WMT', 'DIS', 'PFE', 'TDOC', 'PLUG', 'V',
                       'IVV', 'QQQ', 'IYW']

        #     symbols = ['PLUG']


            for symbol in symbols:
                driver.find_element_by_xpath("//input[@name = 'symbol']").click()
                driver.find_element_by_xpath("//input[@name = 'symbol']").clear()
                driver.find_element_by_xpath("//input[@name = 'symbol']").send_keys(symbol)
                driver.find_element_by_xpath("//input[@value = 'Go']").click()
                time.sleep(4)
                driver.find_element_by_xpath("//span[@class=\"export-menu-triangle\"]").click()
                driver.find_element_by_xpath("//div[@ng-click=\"clickExportSpreadsheet()\"]/..").click()

                time.sleep(5)

                list_of_files = glob.glob(r"C:\Users\yhfdidi\Downloads\Chart*.csv") # * means all if need specific format then *.csv
                latest_file = max(list_of_files, key=os.path.getctime)
                print(symbol + " record file: " + latest_file)

                df = pd.read_csv(latest_file)
                filt = [not pd.isnull(x) for x in df['Open']]
                df_need = df[filt].iloc[-5:]

                today = df_need.iloc[-1]
                if today['Close'] > today['BBUpper']:
                    msg = symbol + " 已突破上布林通道，考虑减仓"
                #     Thread(target = alert_popup, args = (msg))
                    print(msg)
                if today['Close'] < today['BBLower']:
                    msg = symbol + " 已跌破下布林通道，考虑加持"
                    print(msg)

                trend = (df_need['Close'].iloc[-1] - df_need['Close'].iloc[-3]) / df_need['Close'].iloc[-3]
                yesterday = df_need.iloc[-2]

                fom = ['SMA(20)', 'SMA(50)', 'SMA(150)']
                close  = 'Close'

                if trend > 0:
                    sign = "涨超"
                    if (today[close] - yesterday[close]) / yesterday[close] > 0.05:
                        print("今日," + symbol + "已经" + sign + " 5%")
                    for f in fom:
                        if today[close] > today[f] and yesterday[close] < today[f]:
                            print("今日," + symbol + "已经" + sign + f)

                else:
                    sign = "跌破"
                    if (-today[close] + yesterday[close]) / yesterday[close] > 0.05:
                        print("今日," + symbol + "已经" + sign + " 5%")
                    for f in fom:
                        if today[close] < today[f] and yesterday[close] > today[f]:
                            print("今日," + symbol + "已经" + sign + f)

            print('')                
            time.sleep(10)
            
run_main()
```

```python
pro1 = multiprocessing.Process(target = run_main)
pro1.start()
time.sleep(5)
pro1.terminate()
```

```python
list_of_files = glob.glob(r"C:\Users\yhfdidi\Downloads\Chart*.csv")
for file in list_of_files:
#     file = r"C:\Users\yhfdidi\Downloads\Chart-20210104-180844.csv"
    df = pd.read_csv(file)
    filt = [not pd.isnull(x) for x in df['Open']]
    df = df[filt]
    res = 0
    stock = 0
    threshold = 0.0
    days_avg = 5
    for i in range(max(1,df.shape[0] - 90), df.shape[0]):
        if pd.isna(df.iloc[i-10]['Histogram']): continue
        hist_curr = df.iloc[i]['Histogram']
        hist_prev = df.iloc[i-1]['Histogram']
        if abs(hist_curr) > threshold:# and sum(df.iloc[i-days_avg:i+1]['Histogram'])/days_avg > 0:
            if hist_curr > 0 and hist_prev < 0 and not stock and df.iloc[i]['EMA(9,Modern)'] > -2:
                stock = 1
                buy = df.iloc[i]['Close']
#                 print(df.iloc[i]['Date'], 'Buy', df.iloc[i][['Close']].values)

            elif hist_curr < 0 and hist_prev > 0 and stock and df.iloc[i]['Close'] > buy: # and df.iloc[i]['EMA(9,Modern)'] < 0:
                stock = 0
                res += df.iloc[i]['Close'] - buy
#                 print(df.iloc[i]['Date'], 'Sell', df.iloc[i][['Close']].values)
#                 print(df.iloc[i]['Date'], "Revenue: ", df.iloc[i]['Close'] - buy)
    print(round(res/df['Close'].to_list()[-1] * 100,2),'%')

```
